```python

!pip install SpeechRecognition


import speech_recognition as sr


def speech_to_text_from_file(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        print(f"Processando o arquivo de áudio: {audio_file}...")
        audio = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio, language="pt-BR")
            print(f"Texto reconhecido: {text}")
            return text
        except sr.UnknownValueError:
            print("Desculpe, não foi possível entender o áudio.")
        except sr.RequestError:
            print("Desculpe, o serviço de reconhecimento de fala está indisponível.")
    return None


from google.colab import files
uploaded = files.upload()

audio_file = list(uploaded.keys())[0]

texto_reconhecido = speech_to_text_from_file(audio_file)

if texto_reconhecido:
    print("Texto reconhecido com sucesso!")
else:
    print("Não foi possível reconhecer o texto.")
```

```python
!pip install gtts wikipedia geopy


import wikipedia
from geopy.geocoders import Nominatim
from gtts import gTTS
from IPython.display import HTML, display
import os
from urllib.parse import quote


def text_to_speech(text, lang='pt'):
    tts = gTTS(text=text, lang=lang)
    tts.save("output.mp3")
    os.system("start output.mp3")
    print("Áudio gerado e reproduzido.")


def abrir_wikipedia(query):
    try:
        result = wikipedia.summary(query, sentences=2)
        print(result)
        text_to_speech(result)
    except wikipedia.exceptions.DisambiguationError:
        text_to_speech("Por favor, seja mais específico.")
    except wikipedia.exceptions.PageError:
        text_to_speech("Desculpe, não encontrei informações sobre isso.")


def abrir_youtube(query):
    # Codifica a query em formato URL
    encoded_query = quote(query)

    youtube_link = f"https://www.youtube.com/results?search_query={encoded_query}"
    print(f"Link do YouTube: {youtube_link}")
    text_to_speech(f"Estou procurando por {query} no YouTube.")


    display(HTML(f'<a href="{youtube_link}" target="_blank">Clique aqui para ver os resultados no YouTube</a>'))


def localizar_farmacia():
    geolocator = Nominatim(user_agent="virtual_assistant")
    location = geolocator.geocode("farmácia")
    if location:
        text_to_speech(f"A farmácia mais próxima está em {location.address}")
        print(f"Endereço da farmácia mais próxima: {location.address}")
    else:
        text_to_speech("Desculpe, não consegui encontrar uma farmácia próxima.")


def processar_comando(comando):
    if "wikipedia" in comando:
        query = comando.replace("wikipedia", "").strip()
        abrir_wikipedia(query)
    elif "youtube" in comando:
        query = comando.replace("youtube", "").strip()
        abrir_youtube(query)
    elif "farmácia" in comando or "farmacia" in comando:
        localizar_farmacia()
    else:
        text_to_speech("Desculpe, não entendi o comando.")
        print("Comando não reconhecido.")


comando = "youtube música relaxante"
if comando:
    processar_comando(comando.lower())
```

```python
!pip install gtts wikipedia geopy


import wikipedia
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
from gtts import gTTS
from IPython.display import HTML, display
import os
from urllib.parse import quote


def text_to_speech(text, lang='pt'):
    tts = gTTS(text=text, lang=lang)
    tts.save("output.mp3")
    os.system("start output.mp3")
    print("Áudio gerado e reproduzido.")

def localizar_farmacia(endereco_usuario):
    geolocator = Nominatim(user_agent="virtual_assistant")


    user_location = geolocator.geocode(endereco_usuario)
    if not user_location:
        text_to_speech("Não consegui localizar o endereço fornecido.")
        print("Endereço não encontrado.")
        return

    print(f"Endereço fornecido: {user_location.address}")
    print(f"Coordenadas: {user_location.latitude}, {user_location.longitude}")


    farmacias = [
        {"nome": "Farmácia Central", "localizacao": (user_location.latitude + 0.01, user_location.longitude + 0.01)},
        {"nome": "Farmácia São João", "localizacao": (user_location.latitude - 0.01, user_location.longitude - 0.01)},
        {"nome": "Farmácia Popular", "localizacao": (user_location.latitude + 0.02, user_location.longitude)},
    ]


    distancias = []
    for farmacia in farmacias:
        distancia = geodesic((user_location.latitude, user_location.longitude), farmacia["localizacao"]).kilometers
        distancias.append({"nome": farmacia["nome"], "distancia": distancia, "localizacao": farmacia["localizacao"]})


    distancias.sort(key=lambda x: x["distancia"])


    if distancias:
        mais_proxima = distancias[0]
        text_to_speech(f"A farmácia mais próxima é {mais_proxima['nome']}, a aproximadamente {mais_proxima['distancia']:.2f} quilômetros.")
        print(f"Farmácia mais próxima: {mais_proxima['nome']}")
        print(f"Localização: {mais_proxima['localizacao']}")
        print(f"Distância: {mais_proxima['distancia']:.2f} km")
    else:
        text_to_speech("Desculpe, não consegui encontrar farmácias próximas.")


def processar_comando(comando):
    if "wikipedia" in comando:
        query = comando.replace("wikipedia", "").strip()
        abrir_wikipedia(query)
    elif "youtube" in comando:
        query = comando.replace("youtube", "").strip()
        abrir_youtube(query)
    elif "farmácia" in comando or "farmacia" in comando:
        endereco = input("Digite sua cidade ou endereço para encontrar farmácias próximas: ")
        localizar_farmacia(endereco)
    else:
        text_to_speech("Desculpe, não entendi o comando.")
        print("Comando não reconhecido.")


comando = "farmácia"
if comando:
    processar_comando(comando.lower())
```
